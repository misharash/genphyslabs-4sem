```python
def olsim(x,y,sx,sy):
	a=0
	for i in range(3):
		w=1/(sy**2+(a*sx)**2)
		Sw=sum(w)
		ax=sum(x*w)/Sw
		ay=sum(y*w)/Sw
		axy=sum(x*y*w)/Sw
		ax2=sum((x**2+sx**2)*w)/Sw
		ay2=sum((y**2+sy**2)*w)/Sw
		a=(axy-ax*ay)/(ax2-ax*ax)
	b=(ax2*ay-ax*axy)/(ax2-ax*ax)
	n=len(x)
	chi2=sum((y-a*x-b)**2/(sy**2+(a*sx)**2))
	s02=n/(n-2)*(ay2-ay*ay-(axy-ax*ay)*(axy-ax*ay)/(ax2-ax*ax))
	sa2=s02/n/(ax2-ax*ax)
	sb2=sa2*ax2
	sa=sqrt(sa2)
	sb=sqrt(sb2)
	return a,b,sa,sb,chi2
```

```python
data1=genfromtxt("exp1.csv",delimiter=",")
m1=data1[0,1:]
nu1=data1[1:,0]
c1=4e-3 #mm
sigma1=1.5 #divisions
Y1=c1*(data1[1:,1:].T-data1[1:,5]).T
sY1=sigma1*c1
len(Y1)
```

```python
arg1=arange(-4.5,4.501,1)
l1=zeros(4)
sl1=zeros(4)
for i in range(0,4):
    errorbar(m1,Y1[i,:],yerr=sY1,fmt=".",label=r"$\nu={0:0.3f}$ МГц".format(nu1[i]))
    if Y1[i,0]!=Y1[i,0]:
        wm1=m1[1:-1]
        wY1=Y1[i,1:-1]
    else:
        wm1=m1
        wY1=Y1[i,:]
    a,b,sa,sb,chi2=olsim(wm1,wY1,array([0]*len(wm1)),sY1)
    print(a,b,sa,sb,chi2)
    plot(arg1,a*arg1+b)
    l1[i],sl1[i]=a,sa
axis((-4.5,4.5,-.65,.65))
xticks(arange(-4,4.01,1))
minorticks_on()
grid(which='both')
legend(loc='lower right')
title("Положения дифракционных максимумов")
xlabel("m")
ylabel("Y(m), мм")
savefig("graph1.pdf")
```

```python
data2=genfromtxt("exp2.csv",delimiter=",")
m2=data2[0,1:]
nu2=data2[1:,0]
c2=6/3.1 #mm
sigma2=.05 #divisions
Y2=c2*data2[1:,1:]
sY2=sigma2*c2
len(Y2)
```

```python
l2=zeros(8)
sl2=zeros(8)
for i in range(0,8):
    a,b,sa,sb,chi2=olsim(m2-5,Y2[i,:],array([0]*len(m2)),sY2)
    print(a,b,sa,sb,chi2)
    l2[i],sl2[i]=a,sa
```

```python
Lambd2,sLambd2=2*l2,2*sl2
errorbar(1/nu2,Lambd2,yerr=sLambd2,fmt='.')
a,b,sa,sb,chi2=olsim(1/nu2,Lambd2,array([0]*len(nu2)),sLambd2)
print(a,b,sa,sb,chi2)
arg2=arange(.5,1.301,0.1)
plot(arg2,a*arg2+b)
axis((.5,1.3,0.7,1.9))
minorticks_on()
grid(which='both')
title("Зависимость длины волны от периода")
xlabel(r"$1/\nu$, мкс")
ylabel(r"$\Lambda$, мм")
savefig("graph2.pdf")
cs2,scs2=a,sa
print(cs2,scs2)
```
