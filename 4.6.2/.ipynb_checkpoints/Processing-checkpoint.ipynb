```python
def olsim(x,y,sx,sy):
	a=0
	for i in range(3):
		w=1/(sy**2+(a*sx)**2)
		Sw=sum(w)
		ax=sum(x*w)/Sw
		ay=sum(y*w)/Sw
		axy=sum(x*y*w)/Sw
		ax2=sum((x**2+sx**2)*w)/Sw
		ay2=sum((y**2+sy**2)*w)/Sw
		a=(axy-ax*ay)/(ax2-ax*ax)
	b=(ax2*ay-ax*axy)/(ax2-ax*ax)
	n=len(x)
	chi2=sum((y-a*x-b)**2/(sy**2+(a*sx)**2))
	s02=n/(n-2)*(ay2-ay*ay-(axy-ax*ay)*(axy-ax*ay)/(ax2-ax*ax))
	sa2=s02/n/(ax2-ax*ax)
	sb2=sa2*ax2
	sa=sqrt(sa2)
	sb=sqrt(sb2)
	return a,b,sa,sb,chi2
```

# Part I Transparation and reflection

```python
xp,Ip=genfromtxt("passing.csv",delimiter=';',skip_header=1).T
len(xp)
```

```python
xr,Ir=genfromtxt("reflecting.csv",delimiter=';',skip_header=1).T
len(xr)
```

```python
x=xp-xp[0]
sx=0.1
sI=2 #100 div, class of accuracy 2; C is neglectable
T=Ip/Ip[0]
T=T[:-1]
sT=sI/Ip[0]
R=Ir/Ir[-1]*(1-T[-1])
R=R[1:]
sR=sI/Ir[-1]*(1-T[-1])
x=x[:-1]
```

```python
errorbar(x,T,xerr=sx,yerr=sT,fmt='.',label="T")
errorbar(x,R,xerr=sx,yerr=sR,fmt='.',label="R")
errorbar(x,T+R,xerr=sx,yerr=sqrt(sT**2+sR**2),fmt='.',label='T+R')
minorticks_on()
grid(which='both')
legend(loc='lower right')
xlabel("$x$, мм")
title("Зависимость коэффициентов прохождения и отражения от ширины зазора")
axis((-0.2,4.2,-0.1,1.1))
tight_layout()
savefig("graphTR.pdf")
```

```python
lnT=log(T)
slnT=sT/T
errorbar(x,lnT,xerr=sx,yerr=slnT,fmt='.')
a,b,sa,sb,chi2=olsim(x[2:],lnT[2:],sx,slnT[2:])
print(a,b,sa,sb,chi2)
plot(x,a*x+b)
Lambda,sLambda=-1/a,sa/a**2
print(Lambda,sLambda)
xlabel("$x$, мм")
ylabel(r"$\ln T$")
title("Зависимость коэффициента прохождения от ширины зазора")
minorticks_on()
grid(which='both')
savefig('graphlnT.pdf')
```

$$ \Lambda = \frac{\lambda_2}{4\pi \sqrt{n^2 \sin^2 \varphi_1 - 1}} $$

```python
c=3e8
nu=36.53e9
snu=.05e9
lambd=c/nu*1e3
slambd=lambd*snu/nu
print(lambd,slambd) #mm
```

```python
n=sqrt(2*((lambd/(4*pi*Lambda))**2+1))
sn=2*(lambd/(4*pi*Lambda))**2*sqrt((slambd/lambd)**2+(sLambda/Lambda)**2)/n
print(n,sn)
```
